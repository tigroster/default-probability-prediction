# Further Inspection of Selected Predictors

## Obtain Information about Defaults

We have to consider that the debts which are not repaid might still default.
Therefore, for the repaid category we cannot consider loans whose status is not "Repaid".

```python
print(f'Unique values:')
for predicted in FUTURE_PREDICTED:
    print(f'{predicted}: {df[predicted].unique()}')
```

```python
status_values = df['Status'].unique()

for status_value in status_values:
    count_is_nan = ((df['DefaultDate'].isna()) & (df['Status'] == status_value)).to_numpy().sum()
    count_not_nan = ((df['DefaultDate'].notna()) & (df['Status'] == status_value)).to_numpy().sum()
    print(f'Count of isNaN x {status_value}: {count_is_nan}')
    print(f'Count of notNaN x {status_value}: {count_not_nan}')
```

Build count matrix:

|         |     No default |      Default |
|         |         is NaN |      not NaN |
|---------|---------------:|-------------:|
| Late    |         19 547 |   **79 588** |
| Current |        108 826 |      **274** |
| Repaid  |    **102 917** |   **32 869** |

In bold are default combinations which can be used as samples for our prediction.
Thus, we have in total **215 648 samples**, of which there is **102 917 repaid** and **112 731 defaults**.

Now, pick only the above-mentioned rows and selected columns. Save to variable **reduced_df**.


```python
mask = (
    (df['DefaultDate'].notna())
    | ((df['DefaultDate'].isna()) & (df['Status'] == 'Repaid'))
)
reduced_df = df[mask]
reduced_df = reduced_df[FUTURE_PREDICTED + PREDICTORS]
reduced_df
```

Add **Default** as predicted value and remove **DefaultDate** and **Status**.

```python
reduced_df.insert(loc=0, column='Default', value=reduced_df['DefaultDate'].notna())
reduced_df = reduced_df[['Default'] + PREDICTORS]
reduced_df
```

```python
repaid_df = reduced_df[reduced_df['Default'] == False]
default_df = reduced_df[reduced_df['Default'] == True]
```
